Remove zero values

```python
b = pd.read_csv('building_inventory.csv',
               na_values={'Square Footage': 0, 
                         'Year Acquired': 0,
                         'Year Constructed': 0, 
                         'Floors':0})
```

```python
b
```

Year Constructed vs Year Acquired graph

```python
ax = b.plot(x="Year Acquired", y="Year Constructed", figsize=(15,15), kind = 'scatter')
ax.set_xlim(1750, 2021)
ax.set_ylim(1750, 2021)
```

Total square footage vs congress district graph

```python
agg_b = b.groupby("Congress Dist")["Square Footage"].sum()
```

```python
agg_b.index
```

```python
agg_b.values
```

```python
plt.style.use('ggplot')

fig, ax = plt.subplots(1,1)

ax.plot(agg_b.index, agg_b.values, marker='.', linewidth=1.0)
ax.set_xlabel('Congress District')
ax.set_ylabel('Total Square Footage')
ax.set_yscale("log")

plt.show()
```

Avg square footage vs congress district graph

```python
divi_b = b.groupby("Congress Dist")["Total Floors"].sum()
```

```python
avg_floor = divi_b.values
```

```python
avg_floor
```

```python
total_sqrt = agg_b.values
```

```python
total_sqrt
```

```python
sqrt_per_floor = np.divide(total_sqrt, avg_floor) #numpy division element wise
```

```python
sqrt_per_floor
```

```python
plt.style.use('ggplot')

fig, ax = plt.subplots(1,1)

ax.plot(agg_b.index, sqrt_per_floor, marker='.', linewidth=1.0)
ax.set_xlabel('Congress District')
ax.set_ylabel('Average Square Footage per floor')

plt.show()

        
```
